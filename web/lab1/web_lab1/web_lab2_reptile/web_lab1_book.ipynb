```python
def reptile(ID,tag):
    url = 'https://book.douban.com/subject/'+str(ID)
    # headers = {
    # 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
    # AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'
    # }
    headers = {'User-Agent': ua}
    cookies = {'cookie': get_cook}
    html_content = requests.get(url = url, headers = headers, proxies = getip())
    if html_content.status_code == 200:
        print("请求成功")
    else:
        print("请求失败，状态码：" + str(html_content.status_code)+"尝试使用cookie")
        html_content = requests.get(url = url, headers = headers, proxies = getip(),cookies = cookies)
        if html_content.status_code == 200:
            print("cookie成功!")
        else:
            print("cookie失败!")
    soup = BeautifulSoup(html_content.text, "html.parser")

    #书籍名称
    book_name_span = soup.find('span', property='v:itemreviewed')
    if(book_name_span):
        book_name = [book_name_span.get_text()]
    else:
        fail_reptile.append(ID)
        return 0
    #作者
    author = []
    author_span = soup.select("#info > span:nth-child(1)")
    others = soup.select("#info > a:nth-child(2)")
    if author_span:
        authors = author_span[0].find_all('a')
        if(authors):
            for a in authors:
                author.extend(a.get_text())
        elif others:
            author.extend(others[0].get_text().replace('\u3000', '').replace('\n', '').replace(' ',''))
        else:
            pass
    else:
        pass
    #出版社信息
    press_span = soup.find('span', class_='pl', string='出版社:')
    if(press_span and press_span.find_next().name == 'a'):
        press = [press_span.find_next('a').text]
    elif press_span:
        press = [press_span.next_sibling.strip()]
    else:
        press = []
    #类型
    genres = []
    if isinstance(tag, str):  # 检查是否是字符串
        tag_list = tag.split(',')
        merged_set = set(tag_list)
        genres = list(merged_set)
    #出品方
    producer_span = soup.find('span', class_='pl', string='出品方:')
    if(producer_span):
        producer = [producer_span.find_next('a').text]
    else:
        producer = []
    #原作名
    origin_title_span = soup.find('span',class_='pl',string = '原作名:')
    if(origin_title_span):
        origin_title = [origin_title_span.next_sibling.strip()]
    else:
        origin_title = []
    #副标题
    subtitle_span = soup.find('span',class_='pl',string = '副标题:')
    if(subtitle_span):
        subtitle = [subtitle_span.next_sibling.strip()]
    else:
        subtitle = []
    #译者
    translator_span = soup.find('span',class_='pl',string = ' 译者')
    if(translator_span):
        translator = [translator_span.find_next('a').text]
    else:
        translator = []
    #出版年
    year_publication_span = soup.find('span',class_='pl',string = '出版年:')
    if(year_publication_span):
        year_publication = [year_publication_span.next_sibling.strip()]
    else:
        year_publication = []
    #页数
    page_sapn = soup.find('span',class_='pl',string = '页数:')
    if(page_sapn):
        page = [page_sapn.next_sibling.strip()]
    else:
        page = []
    #定价
    price_span = soup.find('span',class_='pl',string = '定价:')
    if(price_span):
        price = [price_span.next_sibling.strip()]
    else:
        price = []
    #装帧
    binding_span = soup.find('span',class_='pl',string = '装帧:')
    if(binding_span):
        binding = [binding_span.next_sibling.strip()]
    else:
        binding = []
    #丛书
    series_books_span = soup.find('span',class_='pl',string = '丛书:')
    if(series_books_span):
        series_books = [series_books_span.find_next('a').text]
    else:
        series_books = []
    #ISBN
    isbn_span = soup.find('span',class_='pl',string = 'ISBN:')
    if(isbn_span):
        isbn = [isbn_span.next_sibling.strip()]
    else:
        isbn = []
    #评分
    rating_element = soup.find('strong', class_='ll rating_num')
    if(rating_element):
        rating_element = [rating_element.string]
    else:
        rating_element = []
    #内容简介
    content_Intro = []
    hid_content = soup.select("#link-report > span.all.hidden")
    short_content = soup.select("#link-report > div")
    if hid_content:
        paragraphs = hid_content[0].find_all('p')
        for paragraph in paragraphs:
            content_Intro.extend(paragraph.get_text().replace('\u3000', '').replace('\n', '').replace(' ',''))
    elif short_content:
        paragraphs = short_content[0].find_all('p')
        for paragraph in paragraphs:
            content_Intro.extend(paragraph.get_text().replace('\u3000', '').replace('\n', '').replace(' ',''))
    else:
        content_Intro = []
    #作者简介
    author_Intro = []
    info = soup.find("div",class_="related_info")
    h2_tag = info.find_all("h2")
    for a in h2_tag:
        span_tag = a.find("span", string="作者简介")
        if(span_tag):
            indent = a.find_next_sibling('div', class_='indent')
            if(indent):
                hid_author_intro = indent.find("span",class_="all hidden")
                no_hid_intro     = indent.find("div",class_="intro")
                if(hid_author_intro):
                    paragraphs = hid_author_intro.find_all('p')
                    for paragraph in paragraphs:
                        author_Intro.extend(paragraph.get_text().replace('\u3000', '').replace('\n', '').replace(' ',''))
                elif no_hid_intro:
                    paragraphs = no_hid_intro.find_all('p')
                    for paragraph in paragraphs:
                        author_Intro.extend(paragraph.get_text().replace('\u3000', '').replace('\n', '').replace(' ',''))
                else:
                    pass
        else:
            pass

    book_data['ID'] = ID
    book_data['作者'] = ' '.join(author)
    book_data['书籍名称'] = ' '.join(book_name)
    book_data['类型'] = ' '.join(genres)
    book_data['出版社'] = ' '.join(press)
    book_data['出品方'] = ' '.join(producer)
    book_data['原作名'] = ' '.join(origin_title)
    book_data['副标题'] = ' '.join(subtitle)
    book_data['译者']   = ' '.join(translator)
    book_data['出版年'] = ' '.join(year_publication)
    book_data['页数'] = ' '.join(page)
    book_data['定价'] = ' '.join(price)
    book_data['装帧'] = ' '.join(binding)
    book_data['丛书'] = ' '.join(series_books)
    book_data['ISBN'] = ' '.join(isbn)
    book_data['评分'] = ' '.join(rating_element)
    book_data['内容简介'] = ' '.join(content_Intro)
    book_data['作者简介'] = ' '.join(author_Intro) 
```

```python
for index, row in book.iterrows():
    if(index<=238):
        continue
    else:
        reptile(row['Id'],row['Tag'])
        book_reptile  = book_reptile._append(book_data,ignore_index=True)
        print (book_data)
```
